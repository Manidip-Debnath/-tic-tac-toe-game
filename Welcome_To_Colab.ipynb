```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Install required packages
!pip install opencv-python-headless numpy matplotlib pillow tqdm

from google.colab import files
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
import zipfile
import shutil

# --- STEP 1: UPLOAD YOUR ZIP FILE ---
print("Please upload a ZIP file containing BMVEC_CELL and HMC3_CELL folders")
uploaded = files.upload()
zip_filename = list(uploaded.keys())[0]

# --- STEP 2: EXTRACT FILES ---
extract_path = '/content/cell_images'
os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"Files extracted to: {extract_path}")

# --- STEP 3: FIND CLASS FOLDERS ---
classes = ["BMVEC_CELL", "HMC3_CELL"]
output_path = '/content/converted_images'

# Search for class folders recursively
class_paths = {}
for root, dirs, files in os.walk(extract_path):
    for cls in classes:
        if cls in dirs and cls not in class_paths:
            class_paths[cls] = os.path.join(root, cls)

# Verify we found the folders
if not class_paths:
    print("ERROR: Could not find BMVEC_CELL or HMC3_CELL folders in the extracted files")
    print("Folder structure in ZIP file:")
    !find "{extract_path}" -type d
    raise Exception("Class folders not found")
else:
    print("Found class folders:")
    for cls, path in class_paths.items():
        print(f"- {cls}: {path}")

# --- STEP 4: PROCESS IMAGES ---
os.makedirs(output_path, exist_ok=True)

def convert_tiff_to_jpg(input_path, output_path):
    try:
        img = cv2.imread(input_path, cv2.IMREAD_ANYCOLOR | cv2.IMREAD_ANYDEPTH)
        if img is None:
            return False

        # Normalize based on bit depth
        if img.dtype == np.uint16:
            img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
        elif img.dtype == np.float32:
            img = (img * 255).astype(np.uint8)
        elif img.dtype != np.uint8:
            min_val, max_val = np.min(img), np.max(img)
            if max_val > min_val:
                img = ((img - min_val) / (max_val - min_val) * 255).astype(np.uint8)
            else:
                img = img.astype(np.uint8)

        # Convert to 3 channels
        if len(img.shape) == 2:
            img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        elif img.shape[2] == 4:
            img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)

        cv2.imwrite(output_path, img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])
        return True
    except Exception as e:
        print(f"Error converting {input_path}: {str(e)}")
        return False

print("\nConverting images...")
for cls, input_dir in class_paths.items():
    output_dir = os.path.join(output_path, cls)
    os.makedirs(output_dir, exist_ok=True)

    # Process all image types
    images = []
    for ext in ('.tif', '.tiff', '.jpg', '.jpeg', '.png'):
        images.extend([f for f in os.listdir(input_dir) if f.lower().endswith(ext)])

    if not images:
        print(f"⚠️ No images found in {input_dir}")
        continue

    print(f"Found {len(images)} images in {cls}")

    for file in tqdm(images, desc=f"Processing {cls}"):
        src = os.path.join(input_dir, file)
        dst = os.path.join(output_dir, os.path.splitext(file)[0] + ".jpg")

        if file.lower().endswith(('.tif', '.tiff')):
            convert_tiff_to_jpg(src, dst)
        else:
            img = cv2.imread(src)
            if img is not None:
                cv2.imwrite(dst, img)

# --- STEP 5: VISUALIZE RESULTS ---
def plot_samples():
    plt.figure(figsize=(15, 8))
    plt.suptitle("Cell Image Samples", fontsize=16)

    for i, cls in enumerate(classes):
        cls_dir = os.path.join(output_path, cls)

        if not os.path.exists(cls_dir):
            print(f"⚠️ No output directory for {cls}")
            continue

        images = [f for f in os.listdir(cls_dir) if f.endswith('.jpg')][:5]

        if not images:
            print(f"⚠️ No images found for {cls}")
            continue

        for j, img_file in enumerate(images):
            img_path = os.path.join(cls_dir, img_file)
            try:
                img = Image.open(img_path)

                plt.subplot(2, 5, i*5 + j + 1)
                plt.imshow(img)
                plt.title(f"{cls}\n{img_file}", fontsize=9)
                plt.axis('off')
            except Exception as e:
                print(f"Error loading {img_path}: {str(e)}")

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

print("\nVisualizing samples...")
plot_samples()

# --- STEP 6: DOWNLOAD RESULTS ---
# Create zip file
shutil.make_archive("/content/converted_images", 'zip', output_path)

# Download
from google.colab import files
files.download("/content/converted_images.zip")
print("\nConversion complete! Downloaded converted_images.zip")
```
